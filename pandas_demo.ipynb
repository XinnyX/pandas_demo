```python
import pandas as pd
import numpy as np

df = pd.DataFrame ({'a' : np.random.randn(6),
                 'b' : ['foo', 'bar'] * 3,
                 'c' : np.random.randn(6)})

df
```

```python
# write the function that you want to apply over the whole dataframe

def add_two(num1):
    return num1 + 2
```

```python
#To apply a function over every row, use the apply method to create a new column
#lambda functions are just functions you don't want to name -- aka anonymous functions

df['Value'] = df.apply(lambda row: add_two(row['a']), axis=1)
df
```

```python
#you can also apply functions conditionally
def add_one(num1):
    return num1 + 1

df['conditional_value'] = df.apply(lambda row: add_two(row['a']) if row['b'] =='foo' else add_one(row['a']), axis=1)
df
```
